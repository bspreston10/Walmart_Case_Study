```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_3 = pd.read_csv('walmart_data.csv')
df_3_female = df_3[df_3['Gender'] == 'F']
```

```python
sorted_df = df_3_female.sort_values('Age')

sns.countplot(data = sorted_df, x='Age', palette='viridis')
```

```python
df_female_age = df_3_female[df_3_female['Age']=='26-35']

sns.countplot(data = df_female_age, x='Product_Category', palette='viridis')
```

Females purchase product 5 the most

```python
plt.figure(figsize=(15,10))
sns.countplot(data = df_3_female, x='Product_Category', hue='Age', palette='viridis', width=0.8)
```
