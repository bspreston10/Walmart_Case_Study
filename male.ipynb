```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df_2 = pd.read_csv('walmart_data.csv')
df_2_male = df_2[df_2['Gender'] == 'M']
df_2_male.head()
```

```python
sorted_df = df_2_male.sort_values('Age')

sns.countplot(data = sorted_df, x='Age', palette='viridis')
```

Since males aged 26=35 make up a majority of the market, lets try and find what they purchase the most

```python
df_male_age = df_2_male[df_2_male['Age']=='26-35']
```

```python
sns.countplot(data = df_male_age, x='Product_Category', palette='viridis')
```

```python
plt.figure(figsize=(15,10))
sns.countplot(data = df_2_male, x='Product_Category', hue='Age', palette='viridis', width=0.8)
```

The two graphs above show that males aged 26-35 purchase product 1, 5, 8 the most. These products should be advertised the most.
