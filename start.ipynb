```python
#Overall View of spending habits for both genders
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('walmart_data.csv')
df.head()
```

```python
sns.histplot(data = df, x = 'Purchase', hue = 'Gender', bins=30, palette='coolwarm')
```

```python
avg_purchase = df.groupby('Gender')['Purchase'].mean().reset_index()
avg_purchase.head()
sns.barplot(data = avg_purchase, x='Gender', y='Purchase', palette='viridis')
plt.ylabel('Average Purchase')
```

These two graphs show that males purchase more and spend more money on Black Friday.

```python
purchase_counts = df.groupby('Marital_Status').size().reset_index(name='Purchase_Count')

purchase_counts['Marital_Status'] = purchase_counts['Marital_Status'].map({0: 'Not Married', 1: 'Married'})


plt.figure(figsize=(10, 6))


sns.barplot(data=purchase_counts, x='Marital_Status', y='Purchase_Count', palette='viridis')


plt.title('Number of Purchases by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Number of Purchases')


plt.show()
```

```python
sorted_df = df.sort_values('Age')

plt.figure(figsize=(15,10))
sns.countplot(data = df, x='Product_Category', hue='Age', palette='viridis', width=0.8)

plt.title('Number of Purchases by Product Category and Age')
plt.xlabel('Product Category')
plt.ylabel('Number of Purchases')


plt.show()
```

The age range of 26-35 buys the most products by far

```python
purchase_counts_city = df.groupby('City_Category').size().reset_index(name='Purchase_Count_City')

purchase_counts_city['City_Category'] = purchase_counts_city['City_Category']


plt.figure(figsize=(10, 6))


sns.barplot(data=purchase_counts_city, x='City_Category', y='Purchase_Count_City', palette='viridis')


plt.title('Number of Purchases by City Category')
plt.xlabel('City Category')
plt.ylabel('Number of Purchases')


plt.show()
```

City category B has the most purchases 

```python
purchase_counts_product = df.groupby('Product_Category').size().reset_index(name='Purchase_Count_Product')

purchase_counts_product['Product_Category'] = purchase_counts_product['Product_Category']


plt.figure(figsize=(10, 6))


sns.barplot(data=purchase_counts_product, x='Product_Category', y='Purchase_Count_Product', palette='viridis')


plt.title('Number of Purchases by Product')
plt.xlabel('Product Type')
plt.ylabel('Number of Purchases')


plt.show()
```

Products 1, 5, and 8 are the most purchased

- Males purchase more and spend more money
- People that are not married purchase more
- Age range of 26-35 purchases the most 
- Products 1, 5, 8 are purchased the most
